```python
import gym 
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
import os
```

```python
environment_name = "CarRacing-v2"
```

```python
env = gym.make(environment_name,)
```

```python
env.action_space
```

```python
env.observation_space
```

```python
obs = env.reset()  # Reset the environment
print(obs)  # Optional: Print the initial observation
env.render()  # Render the environment
```

```python
episodes = 5
for episode in range(1, episodes + 1):
    obs = env.reset()
    done = False
    score = 0 
    
    while not done:
        env.render()  # Render the environment without specifying a render mode
        step_result = env.step(env.action_space.sample())  # Store the entire step result
        obs, reward, done, info = step_result[:4]  # Unpack only the first four values
        score += reward
    
    print('Episode:{} Score:{}'.format(episode, score))
    
env.close()
```

```python
env = gym.make(environment_name)
env = DummyVecEnv([lambda:env])
```

```python
log_path = os.path.join('Training','Logs')
model = PPO('CnnPolicy',env,verbose=1,tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=100000)
```

```python
ppo_path = os.path.join('Training','Saved Models','PPO_Driving_Model')
```

```python
model.save(ppo_path)
```

```python
del model
```

```python
model = PPO.load(ppo_path,env)
```

```python
evaluate_policy(model,env,n_eval_episodes=10,render = True)
```
